Distribution Plots

1. distplot
+ jointplot
+ pairplot
+ rugplot
+ kdeplot

```python
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
tips = pd.read_csv('data/seaborn/tips.csv')
```

```python
tips.head()
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.distplot(tips['total_bill'], kde=True, bins=60)
```

Jointplot (kind)
1. scatter
+ hex
+ reg

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

rugplot

```python
sns.rugplot(tips['total_bill'])
```

Categorical Data Plots

1. factorplot
+ boxplot
+ violinplot
+ stripplot
+ swarmplot
+ barplot
+ countplot

barplot and countplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
len(tips)
```

boxplot and violinplot

```python
sns.boxplot(x='day',y='total_bill', data=tips, palette='rainbow')
```

```python
sns.boxplot(data=tips, palette='rainbow',orient='h')
```

```python
sns.boxplot(x='day',y='total_bill', hue='smoker',data=tips, palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill', hue='sex',data=tips, palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill', hue='sex',data=tips, split=True, palette='coolwarm')
```

stripplot and swarmplot

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex', palette='Set1')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex', palette='Set1', split=True)
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips)
```

factorplot

```python
sns.factorplot(x='sex',y='total_bill', data=tips, kind='bar')
```
